# <font color='black'>Data Pre-Processing</font>

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



df = pd.read_csv(r'C:\Users\karee\Downloads\movies.csv\movies.csv')
```

# <font color='black'>Load Our DataSet</font>


```python
df.head()
```

# <font color='black'>Checking For Null Percentage In Each Column</font>


```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

# <font color='black'>Data Types Of Each Column</font>


```python
df.dtypes
```

# <font color='black'>Removing Null Values</font>


```python
df = df.dropna()
```

# <font color='black'>Changing Data Types Of Columns</font>


```python
df['budget']=df['budget'].astype('long')

df['gross']=df['gross'].astype('long')

df['votes']=df['votes'].astype('long')
```

# <font color='black'>Showing Our Dataset After The Change</font>


```python
df.head()
```

# <font color='black'>Getting the year column to match the released year</font>


```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

# <font color='black'>Sorting the data set according to the highest gross earnings</font>


```python
df = df.sort_values(by:=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

# <font color='black'>Getting Rid Of Duplicates</font>

```python
df=df.drop_duplicates().sort_values(by, ascending=False)
```

# <font color='black'>Showing Our DataSet After The Change</font>

```python
df.head()
```

## <font color='black'>Showing Correlation Between Budget and Gross Earnings Through Scatter Plot</font>

```python
plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
sns.regplot(x= "budget", y="gross", data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'black'})
```

```python
df.corr(method='pearson') 
```

##### <font color='black'>It appears after using pearson method that the correlation between the budget and the gross earnings is the highest correlation</font>

### <font color='black'>Representing Our Data Using Confusion Matrix</font>

```python
correlation_matrix=df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric For Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

# <font color='black'>Numerizing Our DataSet</font>

```python
df_numerized=df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
#Correlation Matrix using numerized data frame
correlation_matrix=df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric For Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

### <font color='black'>Showing The Correlation Between Our Columns In The DataSet After Numerizing The DataSet</font>

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
